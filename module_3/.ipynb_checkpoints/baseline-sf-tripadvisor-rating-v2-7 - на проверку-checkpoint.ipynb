# FUNCTIONS

```python
#Функция разделения строки в список
def str_to_list(string):
    str_list = string.str.split(',')
    return str_list
```

```python
#функция очистки элементов списка от скобок и кавычек
def strip_string_in_list(lists):
    strip_list = []
    for i in lists:
        x = i.strip("'[] ")
        strip_list.append(x)
    return strip_list
```

## 1. Обработка NAN 
У наличия пропусков могут быть разные причины, но пропуски нужно либо заполнить, либо исключить из набора полностью. Но с пропусками нужно быть внимательным, **даже отсутствие информации может быть важным признаком!**   
По этому перед обработкой NAN лучше вынести информацию о наличии пропуска как отдельный признак 

```python
# Для примера я возьму столбец Number of Reviews
data['Number_of_Reviews_isNAN'] = pd.isna(data['Number of Reviews']).astype('uint8')
```

```python
# Далее заполняем пропуски 0, вы можете попробовать заполнением средним или средним по городу и тд...
data['Number of Reviews'].fillna(0, inplace=True)
```

```python
#сделаем столбец с пропусками Cuisine style
data['Cuisine_Style_isNAN'] = pd.isna(data['Cuisine Style']).astype('uint8')
```

```python
#Заполнение Cusine Style самыми популярными по датасету (хуже результат)
#most_cuisine = "['Vegetarian Friendly', 'European', 'Mediterranean']" определены в прошлых версиях
#data['Cuisine Style'].fillna(most_cuisine, inplace=True)
```

```python
#Заполним отдельным названием
lac_cuisine = 'Not_specialized'
data['Cuisine Style'].fillna(lac_cuisine, inplace=True)
```

```python
#сделаем столбец с пропусками Price Range
data['Price_Range_isNAN'] = pd.isna(data['Price Range']).astype('uint8')
```

```python
#Заполним пропуски Price Range модой хуже результат
#data['Price Range'].fillna(data['Price Range'].mode()[0], inplace=True)
```

### 2. Обработка признаков
Для начала посмотрим какие признаки у нас могут быть категориальными.

```python
data.nunique(dropna=False)
```

Какие признаки можно считать категориальными?

Для кодирования категориальных признаков есть множество подходов:
* Label Encoding
* One-Hot Encoding
* Target Encoding
* Hashing

Выбор кодирования зависит от признака и выбраной модели.
Не будем сейчас сильно погружаться в эту тематику, давайте посмотрим лучше пример с One-Hot Encoding:
![](https://i.imgur.com/mtimFxh.png)

```python
#Обработка Cusine style
#Разделим на список и почистим
data['Cuisine Style'] = str_to_list(data['Cuisine Style'])
data['Cuisine Style'] = data['Cuisine Style'].apply(strip_string_in_list)

#Сделаем столбец с количеством кухонь представленных в ресторане
data['Number_of_cuisine'] = data['Cuisine Style'].apply(lambda x: len(x))
```

```python
#Обработаем City и сделаем новые колонки со статистикой по городам

#Сделаем вспомогательный датасет со списком всех кухонь по городам
data_city = pd.DataFrame(data.groupby(['City'])['Cuisine Style'].sum())
#Сделаем колонку с максимальным рангом ресторана по городу(педполагаемое количество ресторанов в городе)
data_city['Max_rank'] = data.groupby(['City']).Ranking.max()
#Добавим колонку м максимальным рангом в основной датасет
data['City_max_rank'] = data['City'].apply(lambda x: data_city['Max_rank'][x])
#общая сумма кухонь по городам
data_city['Len'] = data_city['Cuisine Style'].apply(lambda x: len(x))
#Общее количество ресторанов в датасете по городам
data_city['rest'] = data.groupby(['City']).Restaurant_id.count()
#Добавим колонку с общим количество ресторанов в датасете
data['Rest_in_city'] = data['City'].apply(lambda x: data_city['rest'][x])
#Добавим вспомогательную колонку с суммой всех кухонь по городам
data['City_len_cuis'] = data['City'].apply(lambda x: data_city['Len'][x])
#Добавим колонку со средним количеством кухонь на ресторан в городе
data['City_mean_cuisine'] = round(data['City_len_cuis']/data['Rest_in_city'])
#Удалим вспомогательную колонку
data.drop(['City_len_cuis'], axis = 1, inplace=True)
#Самая частовстречающаяся кухня в городе
data_city['Most_com_cuisine'] = data_city['Cuisine Style'].apply(lambda x: Counter(x).most_common(1)[0][0])
#Добавим в датасет самую часто встречающуюся кухню
data['Most_com_cuisine'] = data['City'].apply(lambda x: data_city['Most_com_cuisine'][x])
#Добавим колонку, есть ли в нашем ресторане самая популярная кухня в городе
data['Most_com_cuisine_in'] = 0

for i in data.index:
    if data['Most_com_cuisine'][i] in data['Cuisine Style'][i]:
        data['Most_com_cuisine_in'][i] = 1
    else: 
        data['Most_com_cuisine_in'][i] = 0

#Нормализуем ранг ресторана в городе, относительно общего количества ресторанов в городе
data['Ranking_norm'] = (data.Ranking - 1)/(data.City_max_rank - 1)
#Посчитаем среднее количество отзывов по городам
data_city['City_mean_reviews'] = data.groupby(['City'])['Number of Reviews'].mean()
#Добавим среднее количество отзывов в городе
data['City_mean_reviews'] = data['City'].apply(lambda x: data_city['City_mean_reviews'][x])
#добавим сколько разных кухонь в каждом городе
data['Num_of_cuisine_in_city'] = data['City'].apply(lambda x: len(set(data_city['Cuisine Style'][x]))) #хуже результат
```

```python
# Label Encoding для Price Range (get_dummies хуже)
label_encoder = LabelEncoder()
data['Price Range'] = pd.Series(label_encoder.fit_transform(data['Price Range']))
```

```python
# Label Encoding для Most_com_cuisine  (get_dummies хуже)
data['Most_com_cuisine'] = pd.Series(label_encoder.fit_transform(data['Most_com_cuisine']))
```

```python
#преобразуем Number of Reviews в долю от кол-ва отзывов всего
Number_of_Reviews_sum = data['Number of Reviews'].sum()
data['Num_of_rev_perc'] = data['Number of Reviews']/Number_of_Reviews_sum
```

```python
#преобразуем Number of Reviews в долю от кол-ва отзывов всего(медленный вариант)
#data['Num_of_rev_perc'] = data['Number of Reviews']
#for i in data.index:
#    data['Num_of_rev_perc'][i] = ((data['Number of Reviews'][i])/(data['Number of Reviews'].sum()))
```

```python
#Обработаем отзывы. Достанем даты отзывов(количество дней, прошедших с 01.01.1970), время между отзывами
#пропуски заполним произольной средней датой(1.1.2016) добавим колонки пропусков в этих колонках

Rewiews_date = data.Reviews.apply(lambda x: x.split('], [')[1].split(','))
data['Rewiews_date_1'] = Rewiews_date.apply(lambda x: x[0].strip("[]' "))
data['Rewiews_date_1isNAN'] = data.Rewiews_date_1.apply(lambda x: 1 if len(x) == 0 else 0).astype('uint8')
data['Rewiews_date_1'] = data.Rewiews_date_1.apply(lambda x: '1/1/2016' if len(x) == 0 else x)
data['Rewiews_date_1'] = data.Rewiews_date_1.apply(lambda x: datetime.strptime(x, '%m/%d/%Y'))
data['Rewiews_date_1'] = data.Rewiews_date_1.apply(lambda x: x.toordinal())
data['Rewiews_date_2'] = Rewiews_date.apply(lambda x: x[-1].strip("[]' "))
data['Rewiews_date_2isNAN'] = data.Rewiews_date_2.apply(lambda x: 1 if len(x) == 0 else 0).astype('uint8')
data['Rewiews_date_2'] = data.Rewiews_date_2.apply(lambda x: '1/1/2016' if len(x) == 0 else x)
data['Rewiews_date_2'] = data.Rewiews_date_2.apply(lambda x: datetime.strptime(x, '%m/%d/%Y'))
data['Rewiews_date_2'] = data.Rewiews_date_2.apply(lambda x: x.toordinal())
data['Rewiews_delta'] = data['Rewiews_date_1'] - data['Rewiews_date_2']
```

```python
#добавим колонку, были ли пропуски в отзывах
data['Rewiews_isNAN'] = data.Reviews.apply(lambda x: 1 if x == '[[], []]' else 0)
```

```python
#ищем дубликаты отзывов
data['Reviews_duple'] = 0
duple_set = set(data[data.Reviews.duplicated()].Reviews)
duple_set.remove('[[], []]')
data['Reviews_duple'] = data.Reviews.apply(lambda x: 1 if x in duple_set else 0)
```

```python
#ищем дубликаты Restaurant_id
data['Restaurant_id_duple'] = 0
duple_set = set(data[data.Restaurant_id.duplicated()].Restaurant_id)
data['Restaurant_id_duple'] = data.Restaurant_id.apply(lambda x: 1 if x in duple_set else 0)
```

```python
#ищем дубликаты URL_TA
data['URL_TA_duple'] = 0
duple_set = set(data[data.URL_TA.duplicated()].URL_TA)
data['URL_TA_duple'] = data.URL_TA.apply(lambda x: 1 if x in duple_set else 0)
```

# Data Preprocessing
Теперь, для удобства и воспроизводимости кода, завернем всю обработку в одну большую функцию.

Не планировал делать класс, для обработки данных, так как времени не было. А так как сейчас можно будет сдавать проект и на каникулах, попробую успеть этот класс написать.

```python
# на всякий случай, заново подгружаем данные
df_train = pd.read_csv('main_task.csv')
df_test = pd.read_csv('kaggle_task.csv')
df_train['sample'] = 1 # помечаем где у нас трейн
df_test['sample'] = 0 # помечаем где у нас тест
df_test['Rating'] = 0 # в тесте у нас нет значения Rating, мы его должны предсказать, по этому пока просто заполняем нулями

data = df_test.append(df_train, sort=False).reset_index(drop=True) # объединяем
data.info()
```

```python
def preproc_data(df_input):
    '''includes several functions to pre-process the predictor data.'''
    
    df_output = df_input.copy()
    
    
    def str_to_list(string):
        str_list = string.str.split(',')
        return str_list


    def strip_string_in_list(lists):
        strip_list = []
        for i in lists:
            x = i.strip("'[] ")
            strip_list.append(x)
        return strip_list
    
    label_encoder = LabelEncoder()
    
    # ################### 1. Предобработка ############################################################## 
    # убираем не нужные для модели признаки
    #df_output.drop(['Restaurant_id','ID_TA', 'URL_TA'], axis = 1, inplace=True)
    
    
    # ################### 2. NAN ############################################################## 
    # Далее заполняем пропуски, вы можете попробовать заполнением средним или средним по городу и тд...
    df_output['Number_of_Reviews_isNAN'] = pd.isna(df_output['Number of Reviews']).astype('uint8')
    df_output['Number of Reviews'].fillna(0, inplace=True)
    
    #вставляем столбец с пропусками Cuisine Style
    df_output['Cuisine_Style_isNAN'] = pd.isna(df_output['Cuisine Style']).astype('uint8')
    #Заполнение отдельным значением
    lac_cuisine = 'Not_specialized'
    df_output['Cuisine Style'].fillna(lac_cuisine, inplace=True)
            
    #сделаем столбец с пропусками Price Range
    df_output['Price_Range_isNAN'] = pd.isna(df_output['Price Range']).astype('uint8')
    #Заполним пропуски Price Range минимальным значением
    df_output['Price Range'].fillna('$', inplace=True)
    
    #Заполним пропуски Reviews модой
    df_output['Reviews'].fillna(df_output['Reviews'].mode()[0], inplace=True)
    
    # ################### 3. Encoding ############################################################## 
    # для One-Hot Encoding в pandas есть готовая функция - get_dummies. Особенно радует параметр dummy_na
    #сделаем дату по городам
    df_output['Cuisine Style'] = str_to_list(df_output['Cuisine Style'])
    df_output['Cuisine Style'] = df_output['Cuisine Style'].apply(strip_string_in_list)
    
    #добавляем количество кухонь в ресторане !!есть зависимость от расположения колонок в датасете!!!
    df_output['Number_of_cuisine'] = df_output['Cuisine Style'].apply(lambda x: len(x))
    
    #Price Range label_encoder
    df_output['Price Range'] = pd.Series(label_encoder.fit_transform(df_output['Price Range']))
    
    # ################### 4. Feature Engineering ####################################################
    
    data_city = pd.DataFrame()
    data_city['Max_rank'] = df_output.groupby(['City']).Ranking.max()
    df_output['City_max_rank'] = df_output['City'].apply(lambda x: data_city['Max_rank'][x])
    data_city = pd.DataFrame(df_output.groupby(['City'])['Cuisine Style'].sum())
    data_city['Max_rank'] = df_output.groupby(['City']).Ranking.max()
    data_city['Len'] = data_city['Cuisine Style'].apply(lambda x: len(x))
    data_city['rest'] = data.groupby(['City']).Restaurant_id.count()
    df_output['Rest_in_city'] = df_output['City'].apply(lambda x: data_city['rest'][x])
    df_output['City_len_cuis'] = df_output['City'].apply(lambda x: data_city['Len'][x])
    df_output['City_mean_cuisine'] = round(df_output['City_len_cuis']/df_output['Rest_in_city'])
    df_output.drop(['City_len_cuis'], axis = 1, inplace=True)
    data_city['Most_com_cuisine'] = data_city['Cuisine Style'].apply(lambda x: Counter(x).most_common(1)[0][0])
    df_output['Most_com_cuisine'] = df_output['City'].apply(lambda x: data_city['Most_com_cuisine'][x])
    df_output['Most_com_cuisine_in'] = 0
    
    for i in df_output.index:
        if df_output['Most_com_cuisine'][i] in df_output['Cuisine Style'][i]:
            df_output['Most_com_cuisine_in'][i] = 1
        else: 
            df_output['Most_com_cuisine_in'][i] = 0
        
    df_output['Ranking_norm'] = (df_output.Ranking - 1)/(df_output.City_max_rank - 1)
    data_city['City_mean_reviews'] = df_output.groupby(['City'])['Number of Reviews'].mean()
    df_output['City_mean_reviews'] = df_output['City'].apply(lambda x: data_city['City_mean_reviews'][x])
    
    #Most_com_cuisine label_encoder
    df_output['Most_com_cuisine'] = pd.Series(label_encoder.fit_transform(df_output['Most_com_cuisine']))
    
    #преобразуем Number of Reviews в долю от кол-ва отзывов всего (важно расположение колонки)
    df_output['Num_of_rev_perc'] = df_output['Number of Reviews']
    
    for i in df_output.index:
        df_output['Num_of_rev_perc'][i] = ((df_output['Number of Reviews'][i])/(df_output['Number of Reviews'].sum()))
    
    #добавляем время первого, второго отзывов и время между ними
    Rewiews_date = df_output.Reviews.apply(lambda x: x.split('], [')[1].split(','))
    df_output['Rewiews_date_1'] = Rewiews_date.apply(lambda x: x[0].strip("[]' "))
    df_output['Rewiews_date_1isNAN'] = df_output.Rewiews_date_1.apply(lambda x: 1 if len(x) == 0 else 0).astype('uint8')
    df_output['Rewiews_date_1'] = df_output.Rewiews_date_1.apply(lambda x: '1/1/2016' if len(x) == 0 else x)
    df_output['Rewiews_date_1'] = df_output.Rewiews_date_1.apply(lambda x: datetime.strptime(x, '%m/%d/%Y'))
    df_output['Rewiews_date_1'] = df_output.Rewiews_date_1.apply(lambda x: x.toordinal())
    df_output['Rewiews_date_2'] = Rewiews_date.apply(lambda x: x[-1].strip("[]' "))
    df_output['Rewiews_date_2isNAN'] = df_output.Rewiews_date_2.apply(lambda x: 1 if len(x) == 0 else 0).astype('uint8')
    df_output['Rewiews_date_2'] = df_output.Rewiews_date_2.apply(lambda x: '1/1/2016' if len(x) == 0 else x)
    df_output['Rewiews_date_2'] = df_output.Rewiews_date_2.apply(lambda x: datetime.strptime(x, '%m/%d/%Y'))
    df_output['Rewiews_date_2'] = df_output.Rewiews_date_2.apply(lambda x: x.toordinal())
    df_output['Rewiews_delta'] = df_output['Rewiews_date_1'] - df_output['Rewiews_date_2']
    
    #добавим колонку, были ли пропуски в отзывах
    df_output['Rewiews_isNAN'] = df_output.Reviews.apply(lambda x: 1 if x == '[[], []]' else 0)

    #ищем дубликаты отзывов
    df_output['Reviews_duple'] = 0
    duple_set = set(df_output[df_output.Reviews.duplicated()].Reviews)
    duple_set.remove('[[], []]')
    df_output['Reviews_duple'] = df_output.Reviews.apply(lambda x: 1 if x in duple_set else 0)
        
    #ищем дубликаты Restaurant_id
    df_output['Restaurant_id_duple'] = 0
    duple_set = set(df_output[df_output.Restaurant_id.duplicated()].Restaurant_id)
    df_output['Restaurant_id_duple'] = df_output.Restaurant_id.apply(lambda x: 1 if x in duple_set else 0)   
    
    #ищем дубликаты URL_TA
    df_output['URL_TA_duple'] = 0
    duple_set = set(df_output[df_output.URL_TA.duplicated()].URL_TA)
    df_output['URL_TA_duple'] = df_output.URL_TA.apply(lambda x: 1 if x in duple_set else 0)
    
    #ищем дубликаты ID_TA
    df_output['ID_TA_duple'] = 0
    duple_set = set(df_output[df_output.ID_TA.duplicated()].ID_TA)
    df_output['ID_TA_duple'] = df_output.ID_TA.apply(lambda x: 1 if x in duple_set else 0)
    
    #City One-Hot Encoding
    df_output = pd.get_dummies(df_output, columns=[ 'City'])
      
    # ################### 5. Clean #################################################### 
    # убираем признаки которые еще не успели обработать, 
    # модель на признаках с dtypes "object" обучаться не будет, просто выберим их и удалим
    object_columns = [s for s in df_output.columns if df_output[s].dtypes == 'object']
    df_output.drop(object_columns, axis = 1, inplace=True)
    df_output.drop(['Number of Reviews'], axis = 1, inplace=True)
    
    return df_output
```

```python
# Импортируем необходимые библиотеки:
from sklearn.ensemble import RandomForestRegressor # инструмент для создания и обучения модели
from sklearn import metrics # инструменты для оценки точности модели
```

```python
# Создаём модель (НАСТРОЙКИ НЕ ТРОГАЕМ)
model = RandomForestRegressor(n_estimators=100, verbose=1, n_jobs=-1, random_state=RANDOM_SEED)
```

```python
# Обучаем модель на тестовом наборе данных
model.fit(X_train, y_train)

# Используем обученную модель для предсказания рейтинга ресторанов в тестовой выборке.
# Предсказанные значения записываем в переменную y_pred
y_pred = model.predict(X_test)
```

```python
# Сравниваем предсказанные значения (y_pred) с реальными (y_test), и смотрим насколько они в среднем отличаются
# Метрика называется Mean Absolute Error (MAE) и показывает среднее отклонение предсказанных значений от фактических.
print('MAE:', metrics.mean_absolute_error(y_test, y_pred))
```
