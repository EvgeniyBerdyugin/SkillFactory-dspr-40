```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from collections import Counter
```

```python
data = pd.read_csv('movie_bd_v5.csv')
data.sample(5)
```

```python
data.describe()
```

```python
#самый большой бюджет
data[data.budget == data.budget.max()]
```

```python
#самый маленький короткометраж
data[data.runtime == data.runtime.max()]
```

```python
data[data.runtime == data.runtime.min()]
```

```python
#средний короткометраж
data.runtime.mean()
```

```python
#медианный короткометраж
data.runtime.median()
```

```python
#самая большая прибыль

#data_profit = data.copy()
data['profit'] = (data.revenue - data.budget)
data[data.profit == data.profit.max()]
```

```python
data[data.profit == data.profit.min()]
```

```python
#количество фильмов со сборами больше бюджета
data[data['profit'] > 0].profit.count()
```

```python
#самый кассовый фильм 2008 года
data_2008 = data[data.release_year == 2008]
data_2008[data_2008.revenue == data_2008.revenue.max()]
```

```python
#самый убыточный фильм с 2012 по 2014 года
data_2012_2014 = data[(data.release_year >= 2012) & (data.release_year <= 2014)]
data_2012_2014[data_2012_2014.profit == data_2012_2014.profit.min()]
```

```python
#самый популярный жанр
def str_to_list(string):
    str_list = string.str.split('|')
    return str_list


data['gen_list'] = str_to_list(data['genres'])
gen_list = data['gen_list'].sum()
Counter(gen_list).most_common(1)

#еще вариант

#data.explode('genres')


#другой вариант
Counter(data.genres.str.split('|').sum()).most_common(1)
```

```python
#самый прибыльный жанр
gen_profit_list = data[data.profit>0].gen_list.sum()
Counter(gen_profit_list).most_common(1)


#другой вариант
Counter(data[data.profit>0].genres.str.split('|').sum()).most_common(1)
```
