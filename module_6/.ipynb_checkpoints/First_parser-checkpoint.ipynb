Парсим по списку ссылок с 0

```python
from bs4 import BeautifulSoup  
import pandas as pd  
import requests
from pprint import pprint
from ast import literal_eval
import re
import time
from datetime import datetime
```

```python
links = pd.read_csv('cars_Total.csv')
```

```python
#Сюда будем сохранять
dff = pd.DataFrame(columns=['car_url', 'info1', 'info2', 'info2owner', 'info2Card', 'info3', 'info4', 'comp_link', 'time'])
```

До 25000

```python
n = 0
for url in links[10000:15000].url:
    try:
        req = requests.get(url, headers={'User-Agent': 'Mozilla/5.0'})
        req.encoding = 'utf-8'
        soup = BeautifulSoup(req.text, 'html.parser')
        dff1 = pd.DataFrame(index=[0], columns=['car_url', 'info1', 'info2', 'info2owner', 'info2Card', 'info3', 'info4', 'comp_link', 'time'])
        dff1.loc[0]['car_url'] = url
        info1 = soup.find_all('script', type="application/ld+json")
        dff1.loc[0]['info1'] = literal_eval(str(info1[0]).strip('<script type="application/ld+json"> </script>'))
        info2 = soup.find_all('div', class_="CardOfferBody__leftColumn")
        dff1.loc[0]['info2'] = soup.find_all('div', class_="CardOfferBody__leftColumn")
        dff1.loc[0]['info2owner'] = info2[0].find_all('div', class_="CardBenefits__item-title")
        dff1.loc[0]['info2Card'] = info2[0].find_all('span', class_="CardInfoRow__cell")
        dff1.loc[0]['info3'] = soup.find_all('section', class_="CardComplectation CardOfferBody__contentIsland")
        info4 = soup.find_all('div', id="sale-data-attributes")
        dff1.loc[0]['info4'] = literal_eval(str(info4[0]).strip('<div data-bem=' 'id="sale-data-attributes" style="display:none"></div>'))
        comp_link = soup.find_all('a', class_="Link SpoilerLink CardCatalogLink SpoilerLink_type_default")
        dff1.loc[0]['comp_link'] = comp_link[0]['href']
        dff1.loc[0]['time'] = datetime.today()
        dff = dff.append(dff1, ignore_index=True)
        n += 1
        if n > 100:
            dff.to_csv('cars_INFO6.csv', index=False)
            n = 0
        print(links[links['url'] == url].index)
    except Exception as e:
            print(e)
```
