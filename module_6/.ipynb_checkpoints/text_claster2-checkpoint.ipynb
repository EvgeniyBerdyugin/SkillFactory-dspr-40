```python
from sentence_transformers import SentenceTransformer
sbert_model = SentenceTransformer('bert-base-nli-mean-tokens')

def calcualte_sentenceBERT_embedding(x):
    try:
        if len(x) == 1:
            sentence_embeddings = sbert_model.encode(x)
        else:
            sentence_embeddings = sbert_model.encode([x])
    except:
        return  None

    return list(sentence_embeddings[0])
```

```python
def bert_embedig_of_list(list_of_sentence):
    i = 0
    emb_list = []
    for sentence in list_of_sentence:
        emb = calcualte_sentenceBERT_embedding(sentence.split(' '))
        emb_list.append(emb)
        if i // 100 == i / 100:
            print(i)
        i += 1
    return emb_list # добавить возврат вектора
```

```python
from sklearn.neighbors import kneighbors_graph
```

```python
# строим матрицу смежности
connectivity = kneighbors_graph(X, n_neighbors=23, include_self=True)
# делаем матрицу смежности симметричной
connectivity = 0.5 * (connectivity + connectivity.T)

from sklearn.cluster import AgglomerativeClustering

ac = AgglomerativeClustering(n_clusters=6,
                             affinity='euclidean', # “euclidean”, “l1”, “l2”, “manhattan”, 
                                                   # “cosine”, or “precomputed”
                             linkage='average', # “ward”, “complete”, “average”, “single”
                             connectivity=connectivity
                            )

ac.fit(X)
y_pred = ac.labels_.astype(np.int)

# Графики полученных кластеров
plt.rcParams['figure.figsize'] = 10, 10
plt.scatter(X[:, 0], X[:, 1], c=y_pred, s=1, alpha=.5)
# plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], c='red', s=200, marker='+')
# plt.show()
```
